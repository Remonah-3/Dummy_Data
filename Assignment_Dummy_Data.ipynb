```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

mean_1 = [-3, 0]
cov_matrix = [[1.0, 0.8],
              [0.8, 1.0]]

data_1 = np.random.multivariate_normal(mean_1, cov_matrix, 500)
print(data_1.shape)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(data_1[:, 0], data_1[:, 1])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Data 1 Scatter Plot')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data_1[:, 0], bins=20, color='blue', alpha=0.7)
plt.title('Histogram of X values')
plt.xlabel('X-axis')
plt.ylabel('Frequency')
```

```python
plt.subplot(1, 2, 2)
plt.hist(data_1[:, 1], bins=20, color='green', alpha=0.7)
plt.title('Histogram of Y values')
plt.xlabel('Y-axis')
plt.ylabel('Frequency')
plt.xlim(-5, 5)

plt.tight_layout()
plt.show()
```

```python
mean_2 = [0, -3]
data_2 = np.random.multivariate_normal(mean_2, cov_matrix, 500)

plt.figure(figsize=(6, 6))
plt.scatter(data_1[:, 0], data_1[:, 1])
plt.scatter(data_2[:, 0], data_2[:, 1])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Both Data Sets')
plt.legend()
plt.show()
```

```python
all_data = np.vstack((data_1, data_2))
print(all_data.shape)
```

```python
labels = np.array([0]*500 + [1]*500).reshape(-1, 1)
labeled_data = np.hstack((all_data, labels))
print(labeled_data.shape)
print(labeled_data[:5])
```
